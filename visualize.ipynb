# Cluster Visualization

```python
# load in the file
import pandas as pd
import numpy

df = pd.read_pickle('2.pkl')
```

```python
import torch
```

```python
df.head()
```

```python
# Extract the numeric values (the outbedding columns)
outbeddings_array = df.iloc[:, :-2].values  # Exclude the 'token' and 'label' columns
```

```python
tokens = df['token']
outbeddings = outbeddings_array
best_labels = df['label']
```

```python
import umap.umap_ as umap
import plotly.express as px
import numpy as np

# Reduce to 2D using UMAP
umap_reducer = umap.UMAP(n_components=2, random_state=42)
reduced_data = umap_reducer.fit_transform(outbeddings)

# Create a DataFrame for visualization
import pandas as pd
data = pd.DataFrame(reduced_data, columns=["UMAP Dimension 1", "UMAP Dimension 2"])
data["Token"] = tokens  # Add tokens
data["Cluster"] = best_labels  # Add cluster labels
data["Index"] = [i for i in range(len(data))]

# Plot with Plotly
fig = px.scatter(
    data,
    x="UMAP Dimension 1",
    y="UMAP Dimension 2",
    color="Cluster",
    text="Token",  # This will show the token on hover
    title="Token Cluster Visualization",
)

# Update traces to show token and index only on hover
fig.update_traces(
    marker=dict(size=8),
    customdata=data[["Token", "Index"]].values,  # Attach tokens and index to customdata
    hovertemplate="<b>Token:</b> %{customdata[0]}<br>"
                  "<b>Index:</b> %{customdata[1]}<br>"
                  "<b>Cluster:</b> %{marker.color}<extra></extra>",  # Custom hover format
    text=None,  # Disable text labels
    hoverinfo="skip",  # Suppress default hover info
)

# Show the interactive plot
fig.show()
```

```python
import plotly.io as pio
pio.write_html(fig, file="cluster_visual.html", auto_open=False)
```
